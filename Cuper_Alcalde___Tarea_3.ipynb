### **Paso 0:** Cargar librerías y seleccionar el conjunto de datos que se utilizará para este assignment.

```python
#Importar Librerias
import pandas as pd
import seaborn as sns
import random
```

La función `cargarDatos()` escogerá al azar el conjunto de datos (dataset) con el que trabajarás en este assignment. Además la función ya te devuelve el dataset cargado en la variable `df`

```python
def cargarDatos():
  datasetList = ["taxis", "flights", "planets"]
  rand_idx = random.randrange(len(datasetList))
  random_dataset = datasetList[rand_idx]
  print("El dataset que utilizarás es: ", random_dataset)
  df = sns.load_dataset(random_dataset)
  return df
```

```python
#Asigna el dataset al azar y carga el dataset
df= cargarDatos()
df.head()
```

### **Ejercicio 1:** Método gráfico de un dato categórico

Analiza de manera gráfica un dato categórico de tu dataset.

```python
if df is not None:
    # Identificar columnas categóricas
    categorical_columns = df.select_dtypes(include=['category', 'object']).columns

    if len(categorical_columns) > 0:
        # Seleccionamos la primera columna categórica para el análisis
        cat_col = categorical_columns[0]
        print(f"Analizando la columna categórica: {cat_col}")

        # Crear un gráfico de conteo
        plt.figure(figsize=(10, 6))
        sns.countplot(data=df, x=cat_col)
        plt.title(f"Distribución de la columna categórica: {cat_col}")
        plt.xticks(rotation=45)
        plt.show()
    else:
        print("No se encontraron columnas categóricas en el dataset.")
else:
    print("No se pudo cargar el dataset.")
```

**La interpretación de tu gráfico anterior:** ------
El grafico nos muestra que se tiene una mayor cantidad de autos color amarillos que verde.

### **Ejercicio 2:** Método gráfico de un dato cuantitativo

Analiza de manera gráfica un dato cuantitativo de tu dataset.

```python
if df is not None:
    # Identificar columnas cuantitativas
    quantitative_columns = df.select_dtypes(include=['number']).columns

    if len(quantitative_columns) > 0:
        # Seleccionamos la primera columna cuantitativa para el análisis
        quant_col = quantitative_columns[0]
        print(f"Analizando la columna cuantitativa: {quant_col}")

        # Crear un histograma
        plt.figure(figsize=(10, 6))
        sns.histplot(data=df, x=quant_col, kde=True)
        plt.title(f"Distribución de la columna cuantitativa: {quant_col}")
        plt.xlabel(quant_col)
        plt.ylabel('Frecuencia')
        plt.show()
    else:
        print("No se encontraron columnas cuantitativas en el dataset.")
else:
    print("No se pudo cargar el dataset.")
```

**La interpretación de tu gráfico anterior:** ------
El grafico nos muestra que hay una mayor frecuencia de viajes con un pasajero, seguido de los viajes con 2 pasajeros.

### **Ejercicio 3**: Método gráfico multivariado para datos cuantitativos


```python
if df is not None:
    # Identificar columnas cuantitativas
    quantitative_columns = df.select_dtypes(include=['number']).columns

    if len(quantitative_columns) > 1:
        print(f"Analizando las columnas cuantitativas: {quantitative_columns.tolist()}")

        # Crear un pairplot para las columnas cuantitativas
        plt.figure(figsize=(12, 8))
        sns.pairplot(df[quantitative_columns])
        plt.suptitle("Pairplot de las columnas cuantitativas", y=1.02)
        plt.show()
    else:
        print("No hay suficientes columnas cuantitativas para un análisis multivariado.")
else:
    print("No se pudo cargar el dataset.")
```

**La interpretación de tu gráfico anterior:** ------
Las relaciones positivas entre distance y fare, y entre fare y tip, sugieren que los costos y las propinas aumentan con la distancia del viaje
